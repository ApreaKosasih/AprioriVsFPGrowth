```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
df = pd.read_excel('datasheet.xlsx')
df.head()
```

```python
df['title'] = df['title'].str.strip()
df.dropna(axis=0, subset=['tid'], inplace=True)
df['tid'] = df['tid'].astype('str')
df = df[~df['tid'].str.contains('C')]
df
```

```python
basket = (df.groupby(['tid', 'title'])['qty']
          .sum().unstack().reset_index().fillna(0)
          .set_index('tid'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
frequent_itemsets = fpgrowth(basket_sets, min_support=0.002, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
